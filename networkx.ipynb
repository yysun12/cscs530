### Using networkx to create a watts strogatz model at baseline 
### The network picks 20% of the edges randomly to remove and then rewire edges at a probablity of 10%


```python
%matplotlib inline
import copy
import networkx as nx
import numpy
import matplotlib.pyplot as plt
import pandas
import random
from networkx.generators.classic import empty_graph, path_graph, complete_graph
import seaborn; seaborn.set()

# Import widget methods
from IPython.html.widgets import *
```

```python
# create a stochastic scale free network with 10 nodes and draw it
ba=nx.barabasi_albert_graph(100,2,seed=40)
ba_layout = nx.spring_layout(ba, iterations=100)
nx.draw_networkx(ba, pos=ba_layout, node_color='#dddddd')

print ba.number_of_edges()
# remove 10% of all edges randomly
# random_item picks one random node

random_item = random.choice(ba.nodes())


# random_item1 picks two random nodes, use random.sample
random_item1 = random.sample(ba.nodes(),2)


# random_edge pikces two random edges
num_edge = int(ba.number_of_edges())

# randomly pick 20% of the edges to remove
remove_edge=int(0.2*num_edge)
random_edge = random.sample(ba.edges(),remove_edge)


# remove the selected edges from the graph
ba.remove_edges_from(random_edge)
#print ba.edges()

nx.draw_networkx(ba, pos=ba_layout, node_color='#dddddd')

# rewire 10% of edges with probability p, set p at 50%
n = int(ba.number_of_nodes())

s = complete_graph(n)

print ba.number_of_edges()

```

```python
p = 0.001
for e in s.edges():
    if random.random() < p:
        ba.add_edge(*e)
ba_layout = nx.spring_layout(ba, iterations=100)
nx.draw_networkx(ba, pos=ba_layout, node_color='#dddddd')
print ba.number_of_edges()
```
